```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x=np.array([11,22,33,44,55])
y=np.array([100,200,300,400,500])
plt.plot(x,y,marker="*",color="green",lw=2,mec='red',mfc='blue',markersize='10')
plt.title('basic line  Plot',fontdict={'color':'blue'})
plt.ylabel('marks')
plt.xlabel('roll no')
plt.show()
```

```python
categories=['A','B','C','D']
values=[23,45,56,90]
plt.bar(categories,values,color='purple')
plt.title('bar chart')
plt.xlabel('categories')
plt.ylabel('Values')
plt.show()
```

```python
qty = np.array ([15,45,60,70,80,33])

fruits = ["Apple", "Banana", "Orange", "Pineapple", "Kiwi", "Mango"]

plt.pie(qty, labels=fruits)

plt.title("Fruits Quantity")

plt.legend()

plt.show()
```
